# model training

```python
X.select_dtypes(include=['object', 'string']).columns
```

```python
X.columns = X.columns.astype(str)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
df = df.rename(columns={
    1: 'Dependents_1',
    2: 'Dependents_2',
    3: 'Dependents_3'
})
```

```python
accuracy = model.score(X_test, y_test) * 100
print(f"Accuracy is {accuracy:.3f}%")
```

```python
from sklearn.tree import DecisionTreeClassifier

model2 = DecisionTreeClassifier(random_state=42)

model2.fit(X_train, y_train)

accuracy = model2.score(X_test, y_test) * 100

print(f"Accuracy is {accuracy:.3f}%")
```

```python
from sklearn.linear_model import LogisticRegression

model3 = LogisticRegression(
    max_iter=5000,
    random_state=42
)

model3.fit(X_train, y_train)

accuracy = model3.score(X_test, y_test) * 100

print(f"Accuracy is {accuracy:.3f}%")
```

# confusion matrics

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
import pandas as pd

results = pd.DataFrame({
    'Model': [
        'Random Forest',
        'Decision Tree',
        'Logistic Regression'
    ],
    'Accuracy (%)': [
        round(model.score(X_test, y_test) * 100, 3),
        round(model2.score(X_test, y_test) * 100, 3),
        round(model3.score(X_test, y_test) * 100, 3)
    ]
})

print(results)
```
